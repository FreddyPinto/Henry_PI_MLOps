```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Cargamos los dataset
def read_datasets(datasets):
    dataframes = {}
    for name in datasets:
        dataframes[name] = pd.read_csv(f'/content/drive/MyDrive/data/{name}.csv', encoding='utf-8', lineterminator='\n')
    return dataframes

datasets = ['df_reviews_sa','user_items', 'steam_games']
dataframes = read_datasets(datasets)

# Convertimos a df.
df_reviews = dataframes['df_reviews_sa']
df_items = dataframes['user_items']
df_games = dataframes['steam_games']
```

```python
# Exploramos df_reviews
df_reviews.describe()
```

```python
# Cantidad
cantidad_df = df_reviews['sentiment_analysis'].value_counts().reset_index()
cantidad_df.columns = ['index', 'cantidad']

# Porcentaje
porcentaje_df = (df_reviews['sentiment_analysis'].value_counts(normalize=True) * 100).reset_index()
porcentaje_df.columns = ['index', 'porcentaje']

# Agregar el símbolo de porcentaje y redondear los valores
porcentaje_df['porcentaje'] = porcentaje_df['porcentaje'].apply(lambda x: "{:.2f}%".format(x))

# Unir los dataframes
df_final = pd.merge(cantidad_df, porcentaje_df, on='index')
df_final
```

```python
# Cantidad
cantidad_df = df_reviews['recommend'].value_counts().reset_index()
cantidad_df.columns = ['index', 'cantidad']

# Porcentaje
porcentaje_df = (df_reviews['recommend'].value_counts(normalize=True) * 100).reset_index()
porcentaje_df.columns = ['index', 'porcentaje']

# Agregar el símbolo de porcentaje y redondear los valores
porcentaje_df['porcentaje'] = porcentaje_df['porcentaje'].apply(lambda x: "{:.2f}%".format(x))

# Unir los dataframes
df_final = pd.merge(cantidad_df, porcentaje_df, on='index')

df_final
```
